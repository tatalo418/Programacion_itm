```python
import pandas as pd

# Leer el archivo Excel con los datos que ya tienes
archivo = 'DATOS_CON_VELOCIDAD.xlsx'  # Ajusta el nombre y la ruta si es necesario
df = pd.read_excel(archivo)

# Calcular la aceleración entre puntos consecutivos
df['ACELERACION'] = df['VELOCIDAD'].diff() / df['TIEMPO'].diff()

# Eliminar las filas con NaN (porque la primera fila no tiene valor anterior)
df = df.dropna()

# Guardar el DataFrame con la nueva columna ACELERACION en un archivo Excel
df.to_excel('DATOS_CON_VELOCIDAD_Y_ACELERACION.xlsx', index=False)

print("El archivo con los resultados ha sido guardado como 'DATOS_CON_VELOCIDAD_Y_ACELERACION.xlsx'.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Leer el archivo con todos los datos
df = pd.read_excel("DATOS_CON_VELOCIDAD_Y_ACELERACION.xlsx")

# Crear las 3 gráficas
plt.figure(figsize=(14, 10))

# 1. Posición vs. Tiempo
plt.subplot(3, 1, 1)
plt.plot(df['TIEMPO'], df['POSICION X'], label='Posición', color='blue')
plt.ylabel("Posición (m)")
plt.title("Posición vs. Tiempo")
plt.grid(True)

# 2. Velocidad vs. Tiempo
plt.subplot(3, 1, 2)
plt.plot(df['TIEMPO'], df['VELOCIDAD'], label='Velocidad', color='green')
plt.ylabel("Velocidad (m/s)")
plt.title("Velocidad vs. Tiempo")
plt.grid(True)

# 3. Aceleración vs. Tiempo
plt.subplot(3, 1, 3)
plt.plot(df['TIEMPO'], df['ACELERACION'], label='Aceleración', color='red')
plt.xlabel("Tiempo (s)")
plt.ylabel("Aceleración (m/s²)")
plt.title("Aceleración vs. Tiempo")
plt.grid(True)

# Mostrar todo junto
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Leer el archivo con velocidad
df = pd.read_excel("DATOS_CON_VELOCIDAD_Y_ACELERACION.xlsx")

# Masa en kg
masa = 0.0395  # 39.5 gramos

# Calcular la energía cinética: (1/2) * m * v^2
df['ENERGIA CINETICA'] = 0.5 * masa * (df['VELOCIDAD'] ** 2)

# Guardar el archivo con la nueva columna
df.to_excel("DATOS_COMPLETOS.xlsx", index=False)

print("✅ Archivo 'DATOS_COMPLETOS.xlsx' generado con energía cinética.")
```

```python
import pandas as pd
import numpy as np

# Cargar el archivo de Excel
archivo = 'DATOS_COMPLETOS.xlsx'  # Reemplaza con la ruta de tu archivo de Excel
df = pd.read_excel(archivo)

# Asegúrate de que la columna de posición X esté presente en el archivo
# Si la columna tiene otro nombre, cámbiala en el siguiente código
posiciones_x = df['POSICION X']  # Reemplaza 'POSICION X' si el nombre de la columna es diferente

# Ángulo de inclinación en grados
angulo_inclinacion = 37

# Convertir el ángulo de grados a radianes
angulo_radianes = np.radians(angulo_inclinacion)

# Calcular la altura en cada punto usando la fórmula h = x * sin(θ)
df['ALTURA'] = posiciones_x * np.sin(angulo_radianes)

# Guardar los resultados en un nuevo archivo Excel
df.to_excel('resultados_con_altura.xlsx', index=False)

# Mostrar el DataFrame con las posiciones y las alturas calculadas
print(df[['POSICION X', 'ALTURA']])
```

```python
import pandas as pd
import numpy as np

# Cargar el archivo de Excel con las posiciones y alturas
archivo = 'resultados_con_altura.xlsx'  # Reemplaza con la ruta de tu archivo de Excel
df = pd.read_excel(archivo)

# Asegúrate de que las columnas necesarias estén presentes en el archivo
# Si las columnas tienen otro nombre, cámbialas en el siguiente código
alturas = df['ALTURA']  # Columna de altura
masa = 0.0395  # Masa del objeto en kilogramos (por ejemplo, 39.5 gramos = 0.0395 kg)
g = 9.81  # Gravedad en m/s^2

# Calcular la energía potencial gravitacional en cada punto usando la fórmula
df['E_PG'] = masa * g * alturas

# Guardar los resultados en un nuevo archivo Excel
df.to_excel('resultados_con_energia_potencial.xlsx', index=False)

# Mostrar el DataFrame con las alturas y las energías potenciales calculadas
print(df[['ALTURA', 'E_PG']])
```
